```python
import os
import random
import mne

def inspect_edf_dataset(data_dir: str, num_samples: int = 3) -> None:
    """
    Randomly selects .edf files from a directory and prints their core metadata.
    
    Args:
        data_dir (str): Path to the directory containing .edf files.
        num_samples (int): Number of random files to inspect.
        
    I/O:
        Input: Directory path.
        Output: Console printout of metadata (sfreq, channels, duration).
        
    Edge Cases:
        - Empty directory or no .edf files.
        - Corrupted .edf files that MNE cannot read.
    """
    if not os.path.exists(data_dir):
        print(f"Error: Directory '{data_dir}' not found.")
        return

    # Gather all .edf files
    files = [f for f in os.listdir(data_dir) if f.endswith('.edf')]
    
    if not files:
        print(f"Error: No .edf files found in '{data_dir}'.")
        return

    print(f"Total .edf files found: {len(files)}")
    
    # Sample randomly to avoid bias from alphabetical ordering
    sample_files = random.sample(files, min(num_samples, len(files)))

    for file in sample_files:
        filepath = os.path.join(data_dir, file)
        print(f"\n{'='*50}")
        print(f"FILE: {file}")
        print(f"{'='*50}")
        
        try:
            # Load headers only (preload=False) to save RAM and time
            raw = mne.io.read_raw_edf(filepath, preload=False, verbose=False)
            
            # Extract Metadata
            sfreq = raw.info['sfreq']
            n_channels = len(raw.ch_names)
            duration_sec = raw.n_times / sfreq
            
            print(f"Sampling Frequency (sfreq) : {sfreq} Hz")
            print(f"Total Duration             : {duration_sec:.2f} seconds ({duration_sec/60:.2f} minutes)")
            print(f"Total Channels             : {n_channels}")
            print(f"\nChannel Names:\n{raw.ch_names}")
            
            # Check for bad channels if pre-marked
            if raw.info['bads']:
                print(f"Pre-marked Bad Channels    : {raw.info['bads']}")
            else:
                print("Pre-marked Bad Channels    : None")
                
        except Exception as e:
            print(f"FAILED to read {file}. File might be corrupted.")
            print(f"Error: {e}")

# --- EXECUTION ---
# Replace with your actual directory path
DATASET_DIR = '/kaggle/input/maaed-raw-dataset'
inspect_edf_dataset(DATASET_DIR, num_samples=3)
```

```python
import os
import mne
import numpy as np
import pywt
from mne.preprocessing import ICA
from mne_icalabel import label_components
from autoreject import AutoReject
import gc
import traceback

# --- 1. WAVELET DENOISER ---
def wavelet_denoise(signal, wavelet='db4', level=4, threshold_scale=0.7):
    coeffs = pywt.wavedec(signal, wavelet, level=level, mode='per')
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    uthresh = threshold_scale * sigma * np.sqrt(2 * np.log(len(signal)))
    coeffs[1:] = [pywt.threshold(c, value=uthresh, mode='soft') for c in coeffs[1:]]
    return pywt.waverec(coeffs, wavelet, mode='per')

def apply_wavelet_to_raw(raw):
    print("    -> Applying Wavelet Denoising...")
    data = raw.get_data()
    denoised_data = np.array([wavelet_denoise(ch) for ch in data])
    
    if denoised_data.shape[1] > raw.n_times:
        denoised_data = denoised_data[:, :raw.n_times]
    elif denoised_data.shape[1] < raw.n_times:
        padding = np.zeros((denoised_data.shape[0], raw.n_times - denoised_data.shape[1]))
        denoised_data = np.hstack([denoised_data, padding])
    
    return mne.io.RawArray(denoised_data, raw.info)

# --- 2. MASTER PREPROCESSING FUNCTION ---
def preprocess_epilepsy_file(file_path, save_dir):
    filename = os.path.basename(file_path)
    subject_id = filename.replace('.edf', '')
    
    try:
        print(f"\nProcessing {filename}...")

        # 1. LOAD DATA
        # Verbose=WARNING hides the annoying EDF annotation/filter warnings
        raw = mne.io.read_raw_edf(file_path, preload=True, verbose='WARNING')
            
        # 2. RENAME & CLEAN CHANNELS (The Fix)
        # Strip 'EEG ' prefix so standard_1020 recognizes them
        raw.rename_channels(lambda x: x.replace('EEG ', ''))
        
        # Drop non-brain channels. A1/A2 are mastoids; dropping them leaves the core 19 channels.
        drop_list = ['EKG', 'MK', 'A1', 'A2']
        raw.drop_channels([c for c in drop_list if c in raw.ch_names], on_missing='ignore')
        
        # ---------------------------------------------------------
        # THE FIX: CORRECT OPERATIONS ORDER
        # ---------------------------------------------------------
        
        # 1. Set Montage
        try:
            raw.set_montage('standard_1020', on_missing='warn')
        except Exception as e:
            print(f"  -> Montage warning: {e}")

        # 2. FILTER FIRST (Kill DC drift and extreme low/high noise)
        # This protects the Average Reference from being hijacked by one bad channel.
        raw.filter(l_freq=1.0, h_freq=45.0, fir_design='firwin', verbose=False)
        
        # 3. AVERAGE REFERENCE SECOND
        # Now it is mathematically safe to average them.
        
        
        # 4. ICA
        print("  -> Running ICA...")
        ica = ICA(n_components=15, method='infomax', fit_params={'extended': True}, random_state=42)
        ica.fit(raw)
        
        ic_labels = label_components(raw, ica, method='iclabel')
        labels = ic_labels['labels']
        
        exclude_idx = [i for i, lab in enumerate(labels) if lab in ['eye blink','muscle','heart','line','channel']]
        ica.exclude = exclude_idx
        print(f"  -> ICA removed {len(exclude_idx)} artifact components.")
        
        raw_clean = ica.apply(raw.copy())
        
        # 5. WAVELET DENOISING
        raw_clean = apply_wavelet_to_raw(raw_clean)

        # 6. EPOCHING
        print("  -> Creating Epochs...")
        epochs = mne.make_fixed_length_epochs(raw_clean, duration=4.0, overlap=1.5, preload=True, verbose=False)
        
        # 7. RELAXED AUTOREJECT
        # Increased n_interpolate to 4 (allows fixing up to 4 bad channels per epoch before dropping)
        print(f"  -> AutoReject checking {len(epochs)} epochs...")
        ar = AutoReject(n_interpolate=[1, 2, 4], consensus=[0.1, 0.2], random_state=42, n_jobs=-1, verbose=False)
        epochs_clean = ar.fit_transform(epochs)

        # ---------------------------------------------------------

        # 8. SAVE
        save_path = os.path.join(save_dir, f"{subject_id}_clean-epo.fif")
        epochs_clean.save(save_path, overwrite=True, verbose=False)
        print(f"✅ Success: Saved {len(epochs_clean)} epochs to {save_path}")

    except Exception as e:
        print(f"❌ Failed {filename}: {e}")
        
    finally:
        if 'raw' in locals(): del raw
        if 'raw_clean' in locals(): del raw_clean
        if 'ica' in locals(): del ica
        if 'epochs' in locals(): del epochs
        if 'epochs_clean' in locals(): del epochs_clean
        gc.collect()

# --- 3. BATCH RUNNER ---
def run_flat_pipeline(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    files = [f for f in os.listdir(input_dir) if f.endswith('.edf')]
    print(f"Found {len(files)} .edf files in {input_dir}")
    
    for file in files:
        file_path = os.path.join(input_dir, file)
        preprocess_epilepsy_file(file_path, output_dir)
        
    print("\n--- Batch Processing Complete ---")

# ==========================================
# EXECUTION
# ==========================================
INPUT_DATASET_DIR = '/kaggle/input/maaed-raw-dataset' 
OUTPUT_SAVE_DIR = '/kaggle/working/MAAED_Cleaned_Epochs'

# Execute
run_flat_pipeline(INPUT_DATASET_DIR, OUTPUT_SAVE_DIR)
```

```python
import os
import mne
import pandas as pd

# ==========================================
# 1. SETUP
# ==========================================
EPOCH_DIR = '/kaggle/working/MAAED_Cleaned_Epochs'

def audit_epoch_dataset(data_dir):
    """
    Scans a directory of MNE Epochs (.fif) and reports subject counts,
    epoch counts, and class balance.
    """
    if not os.path.exists(data_dir):
        print(f"Error: Directory {data_dir} does not exist.")
        return
        
    files = [f for f in os.listdir(data_dir) if f.endswith('-epo.fif')]
    if not files:
        print(f"Error: No .fif files found in {data_dir}.")
        return

    print(f"Scanning {len(files)} files...\n")
    
    audit_data = []
    
    # ==========================================
    # 2. EXTRACT METADATA
    # ==========================================
    for file in files:
        filepath = os.path.join(data_dir, file)
        
        # Extract Subject ID
        subject_id = file.replace('_clean-epo.fif', '')
        
        # Determine Class based on filename
        label = 1 if 'positive' in subject_id else 0
        gender = 'female' if 'female' in subject_id else 'male'
        
        try:
            # preload=False keeps it fast and memory-efficient
            epochs = mne.read_epochs(filepath, preload=False, verbose='WARNING')
            num_epochs = len(epochs)
            
            audit_data.append({
                'Subject_ID': subject_id,
                'Gender': gender,
                'Class': 'Positive (Seizure)' if label == 1 else 'Negative (Normal)',
                'Label': label,
                'Epochs_Retained': num_epochs
            })
            
        except Exception as e:
            print(f"FAILED to read {file}: {e}")

    # ==========================================
    # 3. ANALYSIS & REPORTING
    # ==========================================
    df = pd.DataFrame(audit_data)
    
    # Sort by class and then epoch count
    df = df.sort_values(by=['Label', 'Epochs_Retained'], ascending=[True, False]).reset_index(drop=True)
    
    print("--- INDIVIDUAL SUBJECT SUMMARY ---")
    # Print max rows to see the full list in Kaggle
    with pd.option_context('display.max_rows', None, 'display.max_columns', None):
        print(df[['Subject_ID', 'Class', 'Epochs_Retained']])
    
    print("\n--- DATASET BALANCE (CRITICAL FOR ML) ---")
    total_subjects = len(df)
    total_epochs = df['Epochs_Retained'].sum()
    
    print(f"Total Subjects : {total_subjects}")
    print(f"Total Epochs   : {total_epochs}")
    print("\nEpochs per Class:")
    
    # Aggregate by class
    class_summary = df.groupby('Class').agg(
        Subjects=('Subject_ID', 'count'),
        Total_Epochs=('Epochs_Retained', 'sum'),
        Mean_Epochs_per_Subject=('Epochs_Retained', 'mean')
    ).round(1)
    
    print(class_summary)
    
    return df

# ==========================================
# EXECUTION
# ==========================================
# Run the audit
df_audit = audit_epoch_dataset(EPOCH_DIR)

# Optional: Save audit to CSV for your records
if df_audit is not None:
    df_audit.to_csv('/kaggle/working/epoch_audit_report.csv', index=False)
```

```python
import os
import mne
import numpy as np
from tqdm import tqdm # Provides a progress bar

# ==========================================
# 1. SETUP
# ==========================================
EPOCH_DIR = '/kaggle/working/MAAED_Cleaned_Epochs'
SAVE_DIR = '/kaggle/working/DL_Ready_Data'
os.makedirs(SAVE_DIR, exist_ok=True)

def aggregate_dataset(data_dir, save_dir):
    files = [f for f in os.listdir(data_dir) if f.endswith('-epo.fif')]
    
    X_list = []
    y_list = []
    
    print(f"Aggregating {len(files)} files into Deep Learning tensors...\n")
    
    # Loop with a progress bar
    for file in tqdm(files, desc="Processing Files"):
        filepath = os.path.join(data_dir, file)
        
        # Determine Label
        label = 1 if 'positive' in file else 0
        
        try:
            # Load epochs into RAM
            epochs = mne.read_epochs(filepath, preload=True, verbose='WARNING')
            
            # Extract underlying Numpy array
            # Shape: (n_epochs, n_channels, n_times)
            data = epochs.get_data() 
            
            # Create a label array matching the number of epochs
            labels = np.full(len(data), label)
            
            X_list.append(data)
            y_list.append(labels)
            
        except Exception as e:
            print(f"\nFailed to process {file}: {e}")

    # ==========================================
    # 2. CONCATENATE & VERIFY
    # ==========================================
    print("\nConcatenating arrays (This requires RAM)...")
    X = np.concatenate(X_list, axis=0)
    y = np.concatenate(y_list, axis=0)
    
    print("\n--- AGGREGATION COMPLETE ---")
    print(f"X shape: {X.shape} (Epochs, Channels, Timepoints)")
    print(f"y shape: {y.shape} (Labels)")
    
    # Optional but recommended sanity check:
    assert X.shape[0] == 46410, f"Expected 46410 epochs, got {X.shape[0]}"
    
    # ==========================================
    # 3. SAVE TO DISK
    # ==========================================
    # We save as standard .npy files.
    # Note: These files will be large (likely 1-3 GB for X_data)
    
    x_path = os.path.join(save_dir, 'X_data.npy')
    y_path = os.path.join(save_dir, 'y_labels.npy')
    
    print(f"\nSaving X to {x_path}...")
    np.save(x_path, X)
    
    print(f"Saving y to {y_path}...")
    np.save(y_path, y)
    
    print("Files successfully saved. Ready for PyTorch/TensorFlow.")

# Execute
aggregate_dataset(EPOCH_DIR, SAVE_DIR)
```
